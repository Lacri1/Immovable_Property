```python
from PublicDataReader import Kbland
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm

font_path = r"C:\Users\Lacri\AppData\Local\Microsoft\Windows\Fonts\NanumMyeongjo-Regular.ttf"
font_prop = fm.FontProperties(fname=font_path, size=14)
plt.rc('font', family=font_prop.get_name())
```

```python
land = Kbland()
params = {
    "월간주간구분코드": "01",
    "매물종별구분": "01",
    "매매전세코드": "01",
    "지역코드": "11",
    "기간": "20",
}
df_land = land.get_price_index(**params)
df_land.tail()
```

```python
df_seoul = df_land[df_land['지역명'] == '서울'].copy()
df_seoul['날짜'] = pd.to_datetime(df_seoul['날짜'])
df_seoul = df_seoul.sort_values(by='날짜')
df_seoul
```

```python
df = land.get_jeonse_price_ratio(**params)
df
```

```python
df_seoul_jeonse = df[df['지역명'] == '서울특별시'].copy()
df_seoul_jeonse['날짜'] = pd.to_datetime(df_seoul_jeonse['날짜'])
df_seoul_jeonse = df_seoul_jeonse.sort_values(by='날짜')

df_seoul_jeonse
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='날짜', y='가격지수', data=df_seoul, label='아파트 가격지수 (서울)')
plt.title('서울 지역의 아파트 가격지수 및 전세가격비율 변화', fontproperties=font_prop)
plt.xlabel('날짜', fontproperties=font_prop)
plt.ylabel('가격지수', fontproperties=font_prop)
plt.xticks(rotation=45, fontproperties=font_prop)

sns.lineplot(x='날짜', y='전세가격비율', data=df_seoul_jeonse, label='전세가격비율 (서울)')

plt.tight_layout()
plt.legend(prop=font_prop)  # 범례 추가
plt.show()
```
